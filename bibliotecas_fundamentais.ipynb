Agora iremos trabalhar com nossos dados, e criarmos uma coluna e inserir o valor que precisamos.

* Iremos criar uma funÃ§Ã£o que pegue a mÃ©dia do nosso tomate, e classifique eles entre grande, mÃ©dio e pequeno.

```python
def categorizar_tomate_media(media: int):
    if media >= 40 and media <= 70:
        return 'Tomate mÃ©dio'
    elif media <= 40:
        return 'Tomate pequeno'
    else:
        return 'Tomate grande'
```

Agora iremos criar a coluna que vai receber esse valor de retorno da nossa funÃ§Ã£o.

```python
df_tomato['Categoria_tomate'] = df_tomato['Average'].apply(categorizar_tomate_media)
df_tomato
```
