* Iremos criar uma função que pegue a média do nosso tomate, e classifique eles entre grande, médio e pequeno.

```python
def categorizar_tomate_media(media: int):
    if media >= 40 and media <= 70:
        return 'Tomate médio'
    elif media <= 40:
        return 'Tomate pequeno'
    else:
        return 'Tomate grande'
```

Agora iremos criar a coluna que vai receber esse valor de retorno da nossa função.

```python
df_tomato['Categoria_tomate'] = df_tomato['Average'].apply(categorizar_tomate_media)
df_tomato
```

```python
filtro = df_tomato['Average'] < 40

df_tomato.loc[filtro]
```

## Scikit-Learn

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
url = 'data\diabetes.csv'

df_diabetes = pd.read_csv(url, sep=',')

df_diabetes
```

O objetivo da base de dados é analisar se, por meio de certas características, é possível identificar se uma mulher possui ou não diabetes.

A primeira coisa que devemos fazer ao trabalhar com machine learn, é separar uma parte da nossa base para treino e outra para teste, iremos utilizar a biblioteca **model_selection**, que automaticamente faz essa sepração dos dados.

Iremos separar nossos dados em X, que será todas as colunas que iremos utilizar para fazer a previsão, e Y(target), que será o alvo que queremos atingir.

```python
x = df_diabetes.drop(['Class variable'], axis=1)

x
```

```python
y = df_diabetes['Class variable']
y
```

Agora iremos separa nossa base, em 70% para treino, e 30% para teste.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

Agora iremos escolher um modelo de classificação supervisionado para utilizarmos.

**Modelo escolhido**: KNN.

Agora precisamos definir os hiperparâmetros do nosso modelo, ou seja, determinar quantos vizinhos mais próximos serão considerados na classificação dos dados.

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

Agora iremos treinar nosso modelo.

```python
knn.fit(x_train, y_train)
```

Agora vamos testar o desempenho do nosso modelo utilizando a métrica de acurácia. A acurácia indica a proporção de previsões corretas feitas pelo modelo em relação ao total de previsões, sendo uma medida útil para avaliar sua eficácia em classificações.

```python
accuracy = knn.score(x_test, y_test)
```
