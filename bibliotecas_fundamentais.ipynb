* Iremos criar uma funÃ§Ã£o que pegue a mÃ©dia do nosso tomate, e classifique eles entre grande, mÃ©dio e pequeno.

```python
def categorizar_tomate_media(media: int):
    if media >= 40 and media <= 70:
        return 'Tomate mÃ©dio'
    elif media <= 40:
        return 'Tomate pequeno'
    else:
        return 'Tomate grande'
```

```python
df_tomato['Categoria_tomate'] = df_tomato['Average'].apply(categorizar_tomate_media)
df_tomato
```

## Scikit-Learn

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
url = 'data\diabetes.csv'

df_diabetes = pd.read_csv(url, sep=',')

df_diabetes
```

O objetivo da base de dados Ã© analisar se, por meio de certas caracterÃ­sticas, Ã© possÃ­vel identificar se uma mulher possui ou nÃ£o diabetes.

A primeira coisa que devemos fazer ao trabalhar com machine learn, Ã© separar uma parte da nossa base para treino e outra para teste, iremos utilizar a biblioteca **model_selection**, que automaticamente faz essa sepraÃ§Ã£o dos dados.

Iremos separar nossos dados em X, que serÃ¡ todas as colunas que iremos utilizar para fazer a previsÃ£o, e Y(target), que serÃ¡ o alvo que queremos atingir.

```python
x = df_diabetes.drop(['Class variable'], axis=1)

x
```

```python
y = df_diabetes['Class variable']
y
```

Agora iremos separa nossa base, em 70% para treino, e 30% para teste.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

Agora iremos escolher um modelo de classificaÃ§Ã£o supervisionado para utilizarmos.

**Modelo escolhido**: KNN.

Agora precisamos definir os hiperparÃ¢metros do nosso modelo, ou seja, determinar quantos vizinhos mais prÃ³ximos serÃ£o considerados na classificaÃ§Ã£o dos dados.

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

Agora iremos treinar nosso modelo.

```python
knn.fit(x_train, y_train)
```

Agora vamos testar o desempenho do nosso modelo utilizando a mÃ©trica de acurÃ¡cia. A acurÃ¡cia indica a proporÃ§Ã£o de previsÃµes corretas feitas pelo modelo em relaÃ§Ã£o ao total de previsÃµes, sendo uma medida Ãºtil para avaliar sua eficÃ¡cia em classificaÃ§Ãµes.

```python
accuracy = knn.score(x_test, y_test)
```
